```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = [1,2,3,4,5, 6, 7, 8, 9]
y = [5.3, 7, 4.8, 7.8, 8.1, 9.8, 7.3, 6.3, 5.6]
y_mean = np.mean(y)
plt.xticks([1,2,3,4, 5, 6, 7, 8, 9],labels=['Jonathan','Peter','Meehir','Allen','Anthony',
                                            'Jason', 'Ian', 'Syed', 'Jen'], size = 15)
plt.ylim(2,10)
plt.title("Presentation Time, TimeSeries", size = 25, weight = 'bold')
plt.ylabel('Time in minutes', size = 15)
plt.style.use('seaborn')
plt.axhline(5, linestyle=':', c='r', linewidth = 3)
plt.axhline(y_mean, c = 'black')
plt.legend(['Max time', ' ACTUAL MEAN'])
plt.plot(x,y, linewidth = 5)
```
