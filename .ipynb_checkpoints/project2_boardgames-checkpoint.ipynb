```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
sns.pairplot(df, height=1.2, aspect=1.25);
```

```python
full_prices = df[df['price_mkt'].notna()]
```

```python
lr = LinearRegression()
X = full_prices[['views', 'geek']]
y= full_prices.numfans
lr.fit(X,y)
```

```python
import numpy as np
```

```python
lr.score(X,y)
```

```python
c1,c2 = lr.coef_
```

```python
game_to_check = np.array([26315,6.6])
```

```python
lr.predict(game_to_check.reshape(1,-1))
```

```python
model = sm.OLS(y, sm.add_constant(X)) 
#here pass in def, not like sk

#Fit
fit = model.fit()

#Print out summary
fit.summary()
```
